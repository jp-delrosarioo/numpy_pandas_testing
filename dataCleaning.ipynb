```python
import numpy as np
import pandas as pd

df = pd.read_csv('messy_data.csv')

df.head()
```

```python
# Fill the missing data instead of delete

fill_data = df.fillna({'Age': df['Age'].mean(), 'Science_Score': df['Science_Score'].mean(), 'Math_Score': df['Math_Score'].mean(), 'Attendance_Percent': df['Attendance_Percent'].mean()} )

fill_data
```

```python
# Delete the duplicate rows

orig_rows = fill_data.drop_duplicates()

orig_rows
```

```python
# instead of float convert to integer

int_conv = orig_rows.astype({'Age': 'int32', 'Math_Score': 'int32', 'Science_Score': 'int32', 'Attendance_Percent': 'int32'})

int_conv
```

```python
# center the text

pd.set_option('display.colheader_justify', 'center')
print(int_conv.to_string(justify='center'))
```
